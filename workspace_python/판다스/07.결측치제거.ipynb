```python
df=df=pandas.read_excel('test2.xlsx')
df

```

```python
# 결측치 제거 (NaN 데이터가 포함된 행 또는 열 삭제)
df.dropna() # inplace=True 를 매개변수로 전달하면 저장됨 
df
```

```python
# dropna()의 매개변수 정보
# dropna는 많이 쓰는 매개변수로 axis, how 가 있다.
# ex) df.dropna(axis=? ,how=? ,implace=True)

# axis = 'index' : NaN 데이터를 기준으로 행을 삭제 (default)
# axis = 'columns' : NaN 데이터를 기준으로 열을 삭제
# how = 'any' : NaN값이 하나라도 있으면 삭제 
# how = 'all' : 모든 값이 NaN이면 삭제 

df.dropna(axis='index',how='any',inplace=True)
df
```
