```python
# DataFrame을 생성하기 위한 기본 데이터 
data = {
  '이름':['김수한무','두루미와','거북이','민규','민지'],
  '학급':['개나리','개나리','개나리','노란','노란'],
  '국어점수':[90,50,60,70,80],
  '영어점수':[20,30,10,80,90],
  '평균':[55.0,40.0,35.0,75.0,85.0],
  '취미':['독서','','','헬스','축구']

}

data
```

```python
import pandas
```

```python
# 데이터 프레임 생성
df=pandas.DataFrame(data)
df
```

```python
# 데이터프레임 생성 시 index 지정
df=pandas.DataFrame(data,index=['1번','2번','3번','4번','5번'])

# 인덱스의 이름 설정
df.index.name='뿡'

df
```

```python
# index 초기화 
# True, False 앞글자 대문자 
# drop=True : index가 지워지는 효과
# drop=True 작성하지 않으면 index가 컬럼으로 변경
# index 초기화를 저장하기위해서는 inplace=True 작성
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
# 새로운 내용을 확인하기 위해 데이터 프레임 새로 생성
df = pandas.DataFrame(data)
df
```

```python
# 컬럼을 index로 지정
# 이름을 index로 지정 
df.set_index('이름',inplace=True)
```

```python
df
```

```python
# 위에서 변경한 index를 다시 초기화 
df.reset_index(inplace=True)
```
