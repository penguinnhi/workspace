```python
import pandas

df=df=pandas.read_excel('test2.xlsx')
df
```

```python
# 정렬은 열 데이터를 기준으로 정렬할 수 있고, index를 기준으로 정렬 가능

# index 추가
index_list=['1번','2번','3번','4번','5번']
df.index=index_list
df.index.name='학번'
df
```

```python
# 열 데이터로 정렬

# 평균 기준 오름차순 정렬 / 내림차순 : ascending=False
df.sort_values('평균',ascending=True)

# 국어점수 기준 내림차순 정렬
df.sort_values('국어점수',ascending=False)
```

```python
# 평균 기준 내림차순 정렬 후, 평균이 같으면 국어점수 기준 내림차순 정렬
df.sort_values(['평균','국어점수'],ascending=[False,False])
```

```python
# 인덱스를 기준으로 정렬
df.sort_index(ascending=True)
```
